# Altering transit_timestamp to datetime object to separate time and date for merging with other datasets

```python
df['datetime_timestamp'] = pd.to_datetime(
    df['transit_timestamp'],
    format='%m/%d/%Y %I:%M:%S %p')
```

```python
df.head()
```

```python
print(df['datetime_timestamp'].dt.time.head())
```

```python
df['date'] = df['datetime_timestamp'].dt.date
df['time'] = df['datetime_timestamp'].dt.time
```
